```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
crime = pd.read_csv('crimeRatesByState2005.csv')
crime.head()
```

```python
crime_states = crime[(crime['state']!='District of Columbia')&(crime['state']!='United States')]
crime_states.reset_index(inplace=True)
crime_states.head()
```

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
```

```python
fig,ax = plt.subplots(figsize=(12,7))

# plot call
ax.scatter(x=crime_states['murder'],y=crime_states['burglary'],c='royalblue',alpha=0.5, edgecolors='gray')

# 2nd order poly trendline
z = np.polyfit(crime_states['murder'],crime_states['burglary'], 2)
p = np.poly1d(z)
ax.plot(crime_states['murder'].sort_values(),p(crime_states['murder'].sort_values()),c="darkgray", lw=2.5)

# lowess model
lowess_model = lowess(crime_states['burglary'].sort_values().values,
                      crime_states['murder'].sort_values().values, is_sorted=True, frac=0.23, it=0)
X,Y = [],[]
for i in lowess_model:
    X.append(i[0])
    Y.append(i[1])
ax.plot(X,Y, c="royalblue", lw=2.5)

# Superposition title and subtitle call
fig.suptitle('Murders vs Burglaries in the United States'.upper(),
             x = 0.027, y=1.0, font='emprint', fontsize=20, fontweight='bold', ha='left')
ax.set_title(font='emprint', fontsize=12,
             label='Positive correlation between murder rate and burglary in the United States\nSource: flowingdata.com',
             y=1.05, x=-0.055, loc='left')

# drop borders
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

# axis formating
ax.set_xticks(range(0,11,2))
ax.set_yticks(range(0,1300,200))
ax.set_xlabel('Murders\nPer 100k pop', fontdict={'font':'emprint', 'fontweight':'normal'})
ax.set_ylabel('Burglaries\nPer 100k pop', fontdict={'font':'emprint', 'fontweight':'normal'})

# grid lines
ax.set_axisbelow(True)
ax.grid(b=True, which='major', axis='both', color='white')

# Background color
ax.set_facecolor('#F5F5F5')
# rect = fig.patch
# rect.set_facecolor('#F5F5F5')

# graph show
fig.tight_layout()
plt.show()
```
